# Part 3: Arithmetic With Variable - Market Scenarios

```python
from typing_extensions import final
from typing import ValuesView
# store values
price_per_bag = 800
quantity = 25

# Calculate revenue
total_revenue = price_per_bag * quantity

# Print result
print("Total revenue is:", total_revenue, "LRD")

# Store Values
today_price = 3200
increase = 150

# Calculate tomorrow price
tomorrow_price = today_price + increase

# print result
print("Tomorrow rise price is:" , tomorrow_price, "LRD")

# Store onion cost
onion_cost = 1800

# Divide among 3 traders
cost_per_trader = onion_cost / 3

# Print result
print("cost per trader is:", cost_per_trader, "LRD")

# Original price
original_price = 2500

# Discount rate
discount_rate = 0.10

# Calculate discount
discount_amount = original_price * discount_rate

# Calculate final price
final_price = original_price - discount_amount

# Print results
print("Discount amount:", discount_amount, "LRD")
print("Final price:", final_price, "LRD")
```
